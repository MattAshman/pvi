```python
def download_datasets(root_dir, dataset_urls):
    
    # Temporary dir for downloading datasets
    tmp_dir = root_dir + '/tmp'
    
    if os.path.exists(tmp_dir):
        shutil.rmtree(tmp_dir)
        
    os.mkdir(tmp_dir)
    
    # Loop over datasets and download them into tmp
    for dataset_name, location in dataset_urls.items():
        
        print(f'Downloading {dataset_name} dataset')
        os.mkdir(f'{tmp_dir}/{dataset_name}')
        
        # Base URL for dataset in UCI, and individual file names to download
        base_url, file_names = location
        
        data = []
        for file_name in file_names:
            
            url = f'{base_url}/{file_name}'
            file_location = f'{tmp_dir}/{dataset_name}/{file_name}'
            urllib.request.urlretrieve(url, file_location)
            
            data.append(np.loadtxt(file_location, dtype=str, delimiter=','))
            
        data = np.concatenate(data, axis=0)
        np.save(f'{tmp_dir}/{dataset_name}/data.npy', data)
```
